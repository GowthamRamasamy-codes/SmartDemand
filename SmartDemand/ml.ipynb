```python


# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract year, month, and day as features
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day

# Prepare the features (X) and target (y)
X = data[['Year', 'Month', 'Day']]
y_apples = data['Apples']
y_bananas = data['Bananas']
y_oranges = data['Oranges']
y_grapes = data['Grapes']

# Split the data into training and testing sets
X_train, X_test, y_train_apples, y_test_apples, y_train_bananas, y_test_bananas, y_train_oranges, y_test_oranges, y_train_grapes, y_test_grapes = train_test_split(X, y_apples, y_bananas, y_oranges, y_grapes, test_size=0.2, random_state=42)

# Create separate models for each fruit
model_apples = LinearRegression()
model_bananas = LinearRegression()
model_oranges = LinearRegression()
model_grapes = LinearRegression()


# Train the models
model_apples.fit(X_train, y_train_apples)
model_bananas.fit(X_train, y_train_bananas)
model_oranges.fit(X_train, y_train_oranges)
model_grapes.fit(X_train, y_train_grapes)


# Predict demand for a specific date (e.g., 2022-02-20)
input_date = datetime(2022, 2, 20)
input_features = [[input_date.year, input_date.month, input_date.day]]

predicted_apples = model_apples.predict(input_features)[0]
predicted_bananas = model_bananas.predict(input_features)[0]
predicted_oranges = model_oranges.predict(input_features)[0]
predicted_grapes = model_grapes.predict(input_features)[0]

print("LinearRegression")
print("Predicted demand for", input_date.date())
print("Apples:", predicted_apples)
print("Bananas:", predicted_bananas)
print("Oranges:", predicted_oranges)
print("Grapes:", predicted_grapes)

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from datetime import datetime

# Load the dataset
data = pd.read_csv('fruit_data.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract year, month, and day as features
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day

# Prepare the features (X) and target (y)
X = data[['Year', 'Month', 'Day']]
y_apples = data['Apples']
y_bananas = data['Bananas']
y_oranges = data['Oranges']
y_grapes = data['Grapes']

# Split the data into training and testing sets
X_train, X_test, y_train_apples, y_test_apples, y_train_bananas, y_test_bananas, y_train_oranges, y_test_oranges, y_train_grapes, y_test_grapes = train_test_split(X, y_apples, y_bananas, y_oranges, y_grapes, test_size=0.2, random_state=42)

# Create separate models for each fruit
model_apples = LinearRegression()
model_bananas = LinearRegression()
model_oranges = LinearRegression()
model_grapes = LinearRegression()

# Train the models
model_apples.fit(X_train, y_train_apples)
model_bananas.fit(X_train, y_train_bananas)
model_oranges.fit(X_train, y_train_oranges)
model_grapes.fit(X_train, y_train_grapes)

# Predict demand for a specific date (e.g., 2022-02-20)
input_date = datetime(2022, 2, 20)
input_features = [[input_date.year, input_date.month, input_date.day]]

predicted_apples = model_apples.predict(input_features)[0]
predicted_bananas = model_bananas.predict(input_features)[0]
predicted_oranges = model_oranges.predict(input_features)[0]
predicted_grapes = model_grapes.predict(input_features)[0]


# Calculate R-squared score for each model
r2_apples = model_apples.score(X_test, y_test_apples)
r2_bananas = model_bananas.score(X_test, y_test_bananas)
r2_oranges = model_oranges.score(X_test, y_test_oranges)
r2_grapes = model_grapes.score(X_test, y_test_grapes)

print("\nR-squared scores:")
print("Apples:", r2_apples * 100, "%")
print("Bananas:", r2_bananas * 100, "%")
print("Oranges:", r2_oranges * 100, "%")
print("Grapes:", r2_grapes * 100, "%")


```
